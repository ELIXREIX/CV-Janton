```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = cv2.imread('particles.tif', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(A, cmap='gray')
```

```python
A = np.uint8((A > 125) * 255)
```

```python
np.unique(A)
```

```python
plt.imshow(A, cmap='gray')
```

```python
strelm = np.ones((3, 3), dtype=np.uint8)
```

```python
Ad = cv2.dilate(A, strelm)
Ae = cv2.erode(A, strelm)
```

```python
plt.subplot(1,3,1)
plt.imshow(A, cmap='gray')
plt.subplot(1,3,2)
plt.imshow(Ad, cmap='gray')
plt.subplot(1,3,3)
plt.imshow(Ae, cmap='gray')
```

```python
from skimage.morphology import dilation, erosion
```

```python
Ad = dilation(A, strelm)
Ae = erosion(A, strelm)
plt.subplot(1,3,1)
plt.imshow(A, cmap='gray')
plt.subplot(1,3,2)
plt.imshow(Ad, cmap='gray')
plt.subplot(1,3,3)
plt.imshow(Ae, cmap='gray')
```

```python
from skimage.measure import label, regionprops
```

```python
Ad_label = label(Ad)
```

```python
plt.imshow(Ad_label, cmap='Set1')
```

```python
Ae_label = label(Ae)
```

```python
plt.imshow(Ae_label, cmap='Set1')
```

```python
np.unique(Ae_label)
```

```python
for r in regionprops(Ad_label):
  print(r.label, r.area)
```
